```python
WORD_RE = re.compile(r"[\.\w]{3,}")

# 
REPLACE = {re.compile(r"-[\d]+[^\w]"):"negnum",
           re.compile(r"[\d]+.[\d]+.[\d]+.[\d]+"):"ipaddr",
           re.compile(r".git"):" ' select *", # replace to bad vector))))
           re.compile(r"/tmp/"):" ' select *", #tmppath should trigger
} 

def preprocess(vec):
    vec = vec.lower()
    for regexp, val in REPLACE.items():
      vec = regexp.sub(val , vec)
    return vec
```

```python

class BOW():
  def __init__(self, feature_set, feature_extraction_func):
    self.set = feature_set
    self.len = len(self.set)
    self.feature_extraction_func = feature_extraction_func
    self.word2idx = {v:i for i,v in enumerate(self.set)}
    self.idx2word = {self.word2idx[w]:w for w in self.word2idx}
    print("Bow created: feature_len={}".format(self.len))

  def text2vec(self, text_vector):
    vec = np.zeros(self.len)
    for w in self.feature_extraction_func(text_vector):
      if w in self.word2idx:
        vec[self.word2idx[w]] += 1
    return vec.reshape((1,-1))
  
  def list2matrix(self, list_of_texts):
    matrix = np.zeros((len(list_of_texts), self.len))
    for i,text in enumerate(list_of_texts):
      for w in self.feature_extraction_func(text):
        if w in self.word2idx:
          matrix[i, self.word2idx[w]] += 1
    matrix[matrix>1] = 1
    return matrix

class NgramBOW(BOW):
  def __init__(self, data, n, start_stop=None):
    extract_features = lambda x: ngrams(preprocess(x), n)
    freq_dict = {}
    for vec in data:
      vec = preprocess(vec)
      for ngram in extract_features(vec):
        freq_dict[ngram] = freq_dict.get(ngram, 0) + 1

    features = list(reversed(sorted(freq_dict.keys(), key=lambda x: freq_dict[x])))
    features = features[:int(len(features)*0.7)]
    
    self.full_features = features
    if start_stop is None:
      features = features[:4500]
    else:
      features = features[start_stop[0]:start_stop[1]]
    super().__init__(features, extract_features)

class WordBOW(BOW):
  def __init__(self, data):
    extract_features = lambda x: list(WORD_RE.findall(preprocess(x)) + list(ngrams(preprocess(x), 1)))
    freq_dict = {}
    for vec in data:
      for ngram in extract_features(vec):
        freq_dict[ngram] = freq_dict.get(ngram, 0) + 1

    features = list(reversed(sorted(freq_dict.keys(), key=lambda x: freq_dict[x])))
    features = features[:int(len(features)*0.7)]
    
    print("WordBOW: full_len = {}".format(len(features)))
    self.full_features = features
    features = features[:3000]
    super().__init__(features, extract_features)
```

```python
##FASTTEXT model

class FTdict():
  def __init__(self):
    #compute ngrams

    #924134 ngam_count
    #822520 word count
    #
    self.WORDS_SIZE = 10**5
    self.NGRAMS_SIZE = 5*10**5
    self.PAD = 1         # skipp count
    self.ngrms_n = [1, 2, 3]
  
  def __len__(self):
    return self.WORDS_SIZE + self.NGRAMS_SIZE + self.PAD 

  def vec2indices(self, vec):
    vec = preprocess(vec)
    indices = set()
    for n in self.ngrms_n:
      for ngram in list(ngrams(vec, n)):
          indices.add(fnv1a_32(''.join(ngram).encode()) % self.NGRAMS_SIZE + self.PAD )
    for word in WORD_RE.findall(vec):
        indices.add(fnv1a_32(word.encode()) % self.WORDS_SIZE + self.NGRAMS_SIZE + self.PAD )
    return list(indices)

ftdict = FTdict()
#print(ftdict.vec2indices('/web/.git/HEAD'))      
      
    


```

```python
from torch.utils.data import Dataset, DataLoader, BatchSampler

class MLWafDataset(Dataset):
    def __init__(self, payloads, labels, ftdict):
        self.samples = payloads
        display(payloads.head())
        self.labels = labels
        self.ftdict = ftdict

    def get_dict_size(self):
      return len(self.ftdict)

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, index):
        sample = self.samples[index]
        label = self.labels[index]

        sample = self.ftdict.vec2indices(sample)
        return (sample, label)


class PaddingCollator():
  def __init__(self):
    self.PAD_VAL = 0

  def __call__(self, batch):
    samples, labels = list(zip(*batch))
    max_len = max(map(len,samples))
    samples = list(map(lambda x: x + [self.PAD_VAL] * (max_len - len(x)), samples))

    return (torch.IntTensor(samples), torch.Tensor(labels))

if 0:
  testset = MLWafDataset(X_test, y_test, ftdict)
  test_loader = DataLoader(testset,  batch_size=32, collate_fn=PaddingCollator())
  samples, labels = iter(test_loader).next()
  print(samples)
  print(labels)
  del testset, test_loader, samples, labels
```

```python

DIM = 8
DEVICE = 'cuda'
class FastTextModel(nn.Module):
  def __init__(self, dict_size):
    super().__init__()
    self.emb = nn.Embedding(dict_size, DIM, padding_idx=0, scale_grad_by_freq=True)
    self.linear = nn.Linear(DIM, 1)
    self.sigmoid = nn.Sigmoid()

    global bad_count, good_count
    self.loss_func = nn.BCEWithLogitsLoss(pos_weight=torch.Tensor([good_count / bad_count]))

    if DEVICE:
      self.to(DEVICE)
    self.optimizer = torch.optim.Adam(self.parameters(), lr=0.01, weight_decay=1e-6)

  
  def forward(self, input):
    #input (BS, max(emb_len)) padded with zeros
    input = self.emb(input)
    input = torch.sum(input, 1) # (BS, DIM)
    output = self.linear(input) #(BS, 1)
    return output.view(-1)

  def fit(self, input, targets):
    self.train()
    self.optimizer.zero_grad()
    
    if DEVICE:
      input = input.to(DEVICE)
      targets = targets.to(DEVICE)

    outputs = self.forward(input)
    #print(outputs.size(), targets.size())
    loss = self.loss_func(outputs, targets)
    predicted = self.sigmoid(outputs)
    
    loss.backward()
    self.optimizer.step()
    return loss.item(), predicted.to('cpu')

  def predict(self, input):
    self.eval()
    with torch.no_grad():
      if DEVICE:
        input = input.to(DEVICE)
      outputs = self.forward(input)
      predicted = self.sigmoid(outputs)
    return predicted.to('cpu') 

  
  def predict_with_loss(self, input, targets):
    self.eval()
    with torch.no_grad():
      if DEVICE:
        input = input.to(DEVICE)
        targets = targets.to(DEVICE)
      outputs = self.forward(input)
      loss = self.loss_func(outputs, targets)
      predicted = self.sigmoid(outputs)
    return loss.item(), predicted.to('cpu')




model = FastTextModel(20)
input = torch.IntTensor([
         [1,2,3,4,0],
         [2,3,4,0,0],
         [9,19,0,0,0],
])
print(model)
output = model.predict(input)
print(output)
targets = torch.FloatTensor([0,1,1])
loss = model.fit(input, targets)
predicted = model.predict(input)
print(predicted)
print( sum((predicted>0.5) == targets))
print(len(targets))
del model, input, output, targets, loss, predicted
```

```python
#train model
epochs = 10
model = FastTextModel(trainset.get_dict_size())
max_accyracy = 0.993
for epoch in range(epochs):
  epoch_start_time = time.time()

  train_losses, test_losses = [], []
  train_count, test_count = 0, 0
  train_right, test_right = 0, 0
  iteration_count = len(train_loader)
  for i, (input, labels) in enumerate(train_loader, 1):
    print("\repoch {}: iteration {}/{} train".format(epoch, i, iteration_count), end='')
    loss_val, predicted = model.fit(input, labels)
    train_right += sum((predicted>0.5) == labels)
    train_count += len(labels)
    train_losses.append(loss_val)
  train_loss = sum(train_losses)/len(train_losses)
  train_accuracy = train_right/train_count

  for i, (input, labels) in enumerate(test_loader, 1):
    print("\repoch {}: iteration {}/{} test".format(epoch, i, iteration_count), end='')
    loss_val, predicted = model.predict_with_loss(input, labels)
    test_right += sum((predicted>0.5) == labels) # prc needed
    test_count += len(labels)
    test_losses.append(loss_val)
  test_loss = sum(test_losses)/len(test_losses)
  test_accuracy = test_right/test_count
  print("\rEpoch {:3d}:".format(epoch), end='')
  print(" train_loss={} train_accuracy={}".format(train_loss, train_accuracy), end='')
  print(" test_loss={} test_accuracy={}".format(test_loss, test_accuracy), end='')
  print(" time: {}s".format(int(time.time()-epoch_start_time)))
  if test_accuracy > max_accyracy:
    torch.save(model, "fastText_{}.pt".format(test_accuracy))
    max_accyracy = test_accuracy

```
